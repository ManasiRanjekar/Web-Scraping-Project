# Page Scraper - Create an application which connects to a site and pulls out the contents list.

#### Step1: Import libraries

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

#### Step 2:Set the url and get the webpage specified in the url.

```python
url="https://en.wikipedia.org/wiki/Python_(programming_language)"
```

```python
wiki=requests.get(url)
```

#### Step 3:use bs4 to format saved webpage.

```python
soup=BeautifulSoup(wiki.content,'html5lib')
```

#### Step 4:Print formatted webpage

```python
#print(soup.prettify())
```

```python
print(soup.find('title'))
```

```python
contents=soup.find_all("div",class_='toc')
for i in contents:
    print(i.text)
```
